```python
import random
import pandas as pd

quiz_list = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/빅분기 실기 단답형 대비.xlsx', )
quiz_list['기출 여부']=quiz_list['기출 여부'].fillna(0)
quiz_list=quiz_list.dropna()
quiz_num = range(quiz_list.shape[0])

def print_quize(i):
  unknown= 0
  if(quiz_list.iloc[i,2] == 1):
    print("-----------기출-----------")
  print('문제', i,)
  print(quiz_list.iloc[i,0])
  x = input()
  if x == '?':
      unknown = 1
  print('정답 : ', quiz_list.iloc[i,1], '\n')
  return(unknown, x)

def study_hard(mode = 'random'):
  unknown = 0
  unknown_quiz = []
  if mode =='random' : 
    while 1 :
      unknown_quiz = []
      selected_quize = random.sample(quiz_num, 10)
      
      for i in range(10):
        q_num = selected_quize[i]
        result, x = print_quize(q_num)
        if x =='x':
          break;
        unknown = unknown + result
        if result == 1:
          unknown_quiz.append(q_num)

      
      if len(unknown_quiz) != 0 :
        print("?수 : ", unknown)
        print("-----------? 문제 ---------")
        for num in unknown_quiz:
          print_quize(num)
      print("\nx를 눌러 조의를 표하시오")
      x = input()
      if x =='x':
        break;

  if mode == 'all':
    unknown_quiz = []
        
    selected_quize = random.sample(quiz_num, len(quiz_list))
    for i in range(len(selected_quize)):
      q_num = selected_quize[i]
      result, x = print_quize(q_num)
      if result == 1:
        unknown_quiz.append(q_num)
      if x =='x':
        break;

    if len(unknown_quiz) != 0 :
      print("?수 : ", unknown)
      print("-----------? 문제 ---------")
      for num in unknown_quiz:
        print_quize(num)    
```

```python
study_hard()
```

```python
study_hard('all')
```
