주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과
2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.
단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
(최종값은 소수점 반올림 후 정수 출력)

```python
# 라이브러리 불러오기
import pandas as pd
df = pd.read_csv('./basic2.csv')
df.head()
```

```python
# a: 2022년 월별 Sales 합계 중 max
# b: 2023년 월별 Sales 합계 중 max
# abs(a-b)
# 단,Event 컬럼 == 1, 80% Sale값 반영
# 최종값: 반올림해서 정수 출력
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
```

```python
df
```

```python
cond1 = df['year'] == 2022
cond2 = df['year'] == 2023
cond3 = df['Events'] == 1
```

```python
df2022_80 = df[ cond1 & cond3 ]
df2022_80
```

```python
df2022_80 = df[ cond1 & cond3 ]
df2023_80 = df[ cond2 & cond3 ]
df2022_100 = df[ cond1 & ~cond3 ]
df2023_100 = df[ cond2 & ~cond3 ]
```

```python
df2022_80['Sales'] = df2022_80['Sales']*0.8
df2023_80['Sales'] = df2023_80['Sales']*0.8
```

```python
df2022 = pd.concat([df2022_80,df2022_100])
df2023 = pd.concat([df2023_80,df2023_100])
```

```python
max_2022 = df2022.groupby('month')['Sales'].sum().max()
max_2023 = df2023.groupby('month')['Sales'].sum().max()
```

```python
answer = int(round(abs(max_2022-max_2023),0))
print(answer)
```
